```python
import pandas as pd

hashtags = pd.read_csv('news_hashtags.csv')
creaters = pd.read_csv('news_creaters.csv')
```

```python
hashtags
hashtags_list = hashtags['Hashtag name'].tolist()
```

```python
creaters
creaters_list = creaters['Username'].tolist()
```

```python
data_to_filter = pd.read_csv('TEST-results.csv') #change this line to parse different files
data_to_filter
```

```python
import re
def getHashtags(phrase):
    if isinstance(phrase, str):
        return re.findall(r'#(\S+)', phrase.lower())
    else:
        return []

data_to_filter['hashtags'] = data_to_filter['video_description'].apply(getHashtags)
data_to_filter
```

```python
newsDF = data_to_filter[data_to_filter.apply(lambda row: 
                      (isinstance(row['hashtags'], list) and 
                       any(hashtag in row['hashtags'] for hashtag in hashtags_list)) 
                      or row['author_username'] in creaters_list, axis=1)]

## if hashtags is indeed a list, check if there is overlap between the video hashtag and the hashtag list or if the video 
## content creator is a news account. If any of the two is satisfied, pass to new dataset newsDF.
```
