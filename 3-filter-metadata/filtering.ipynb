```python
data_to_filter = data_to_filter.loc[data_to_filter['video_is_ad'] == False]
```

```python
import re
def getHashtags(phrase):
    if isinstance(phrase, str):
        return re.findall(r'#(\S+)', phrase.lower())
    else:
        return []

data_to_filter['hashtags'] = data_to_filter['video_description'].apply(getHashtags)

def getSuggestedWords(phrase):
    if phrase is None:
        return []
    elif isinstance(phrase, str):
        return phrase.split(',')
    else:
        return []
    
data_to_filter['suggested_words'] = data_to_filter['suggested_words'].apply(getSuggestedWords)
```

```python
data_to_filter
```

```python
def filter_conditions(row):
    has_hashtag = isinstance(row['hashtags'], list) and any(hashtag in row['hashtags'] for hashtag in hashtags_list)
    is_creator = row['author_username'] in creaters_list
    has_suggested_word = isinstance(row['suggested_words'], list) and any(word in row['suggested_words'] for word in hashtags_list)
    
    return has_hashtag or is_creator or has_suggested_word

data_to_filter_result = data_to_filter[data_to_filter.apply(filter_conditions, axis=1)]

## if hashtags is indeed a list, check if there is overlap between the video hashtag and the hashtag list or if the video 
## content creator is a news account. If any of the two is satisfied, pass to new dataset newsDF.
```
